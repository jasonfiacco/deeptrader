## Label the Data with Sentiment Rating and Save to CSV

```python
import ssl
import pandas as pd
import ta
from datetime import datetime, timedelta

ssl._create_default_https_context = ssl._create_unverified_context # Only used if pandas gives a SSLError

def label_data(symbol, sentiment_keyword=None, add_ta=False):
    path = "/Users/jasonfiacco/Documents/Yale/Senior/thesis/predictit_datasets/"
    filename = "{}.xlsx".format(symbol)



    df = pd.read_excel(path + filename, skiprows=4)
    df = df.set_index("Date")
    df = df.drop(df.columns[[7,8,9]], axis=1)
    df = df.drop("ID", 1)
    
    
    #Get the normalized price_norm column
    df['price_norm'] = df['Close']/df['Open']
    
    
    #Slice a specific date.
    df.index = pd.to_datetime(df.index)
    df = df[(df.index >= '02-01-2017') & (df.index <= '02-01-2019')]

    #Fill all NaN's for PRICE NORM with just 1 (based on literature)
    #df = df.bfill(axis="rows")
    df[['Open', 'High', 'Low', 'Close', 'Volume']] = df[['Open', 'High', 'Low', 'Close', 'Volume']].bfill(axis="rows")
    df[["price_norm"]] = df[["price_norm"]].fillna(value=1)

    
    #Get Sentiment Analysis
    #First convert date column to appropriate format
    df['current_date'] = pd.to_datetime(df.index)
    df['next_date'] = df['current_date']  + timedelta(days=1)
    df['current_date']  = df['current_date'].dt.strftime("%Y-%m-%d")
    df['next_date'] = df['next_date'].dt.strftime("%Y-%m-%d")
    
    df['sentiment'] = df.apply(lambda row: analyzer.get_sentiment_for_keyword(sentiment_keyword, row['current_date'], row['next_date']), axis=1)
    #df['sentiment'] = vget_sentiment_for_keyword(sentiment_keyword, df['current_date'], df['next_date'])
    del df['next_date']
    
    
    #Rename
    df.columns = [symbol + ":" + name.lower() for name in df.columns]
    
    #Add all the TA extra features
    if add_ta:
        ta.add_all_ta_features(
            df, 
            colprefix=symbol+":",
            **{k: symbol+":" + k for k in ['open', 'high', 'low', 'close', 'volume']}
        )
        
        
    #Now Save the df to a CSV
    df.to_csv(path+symbol+"_labeled.csv")

    return df
```

```python
#Already labeled
label_data("WARREN", sentiment_keyword="elizabeth warren", add_ta=True)
label_data("MANCHIN", sentiment_keyword="joe manchin", add_ta=True)
label_data("SANDERS", sentiment_keyword="bernie sanders", add_ta=True)
label_data("NELSON", sentiment_keyword="bill nelson", add_ta=True)
label_data("CRUZ", sentiment_keyword="ted cruz", add_ta=True)
label_data("DONNELLY", sentiment_keyword="joe donnelly", add_ta=True)
label_data("PELOSI", sentiment_keyword="nancy pelosi", add_ta=True)
label_data("MANAFORT", sentiment_keyword="paul manafort", add_ta=True)
label_data("BROWN", sentiment_keyword="sherrod brown", add_ta=True)
label_data("RYAN", sentiment_keyword="paul ryan", add_ta=True)
label_data("STABENOW", sentiment_keyword="debbie stabenow", add_ta=True)
```
